```python
import requests
from bs4 import BeautifulSoup

def scrape_title(url):
  response = requests.get(url)
  soup = BeautifulSoup(response.content, 'html.parser')
  title = soup.find('title').text
  return title

print(scrape_title('https://www.google.com'))
```
